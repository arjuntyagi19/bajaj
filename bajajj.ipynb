```python
def extract_first_name(row):
  if isinstance(row, dict) and 'patientDetails' in row and isinstance(row['patientDetails'], dict) and 'firstName' in row['patientDetails']:
    return row['patientDetails']['firstName']
  else:
    return None

# Assuming your data is in a list called 'data'
first_names = [extract_first_name(row) for row in df]

print(first_names)
```

```python
def extract_patient_details(patient_details, key):
  if isinstance(patient_details, dict) and key in patient_details:
    return patient_details.get(key)
  else:
    return None

df['firstName'] = df['patientDetails'].apply(lambda x: extract_patient_details(x, 'firstName'))
df['lastName'] = df['patientDetails'].apply(lambda x: extract_patient_details(x, 'lastName'))
df['DOB'] = df['patientDetails'].apply(lambda x: extract_patient_details(x, 'birthDate'))
```

```python

age_group_counts = {
    'Child': 0,
    'Teen': 0,
    'Adult': 0,
    'Senior': 0,
    'Unknown': 0
}

for appointment in df:
    age_group = appointment['patientDetails'].get('ageGroup', 'Unknown')
    if age_group:
        age_group_counts[age_group] += 1

age_group_counts['Adult']
```

```python
def extract_birthdate(patientDetails):
  if isinstance(patientDetails, dict) and 'birthDate' in patientDetails:
    return patientDetails['birthDate']
  else:
    return None

df['birthDate'] = df['patientDetails'].apply(extract_birthdate)
```

```python
def is_valid_indian_number(number):

  if not isinstance(number, (int, float)):
    return False
  number_str = str(int(number))
  if number_str.startswith(('91', '+91')) and len(number_str) == 12 and 6000000000 <= int(number_str[2:]) <= 9999999999:
    return True
  return False


df['isValidMobile'] = df['phoneNumber'].apply(is_valid_indian_number)
```
